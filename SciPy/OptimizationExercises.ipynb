```python
t = np.arange(0.,80.,1.0)
waveform=np.array([ 3.,  2.,  2.,  2.,  2.,  2.,  1.,  2.,  
                   2.,  2.,  2.,  2.,  2.,  5., 10., 18., 25., 25., 20., 13., 
                   9.,  9., 11., 12., 12., 10.,  8.,  8.,  8.,  8.,  7.,  6.,  5., 
                   3.,  3.,  3.,  3.,  2.,  2.,  3.,  3.,  4., 3.,  3.,  4.,  2.,  3.,  
                   4.,  3.,  3.,  3.,  3.,  3.,  3.,  2.,  3.,  3.,  3.,  2.,  2.,  2., 
                   1.,  2.,  3.,  3., 3.,  2.,  2.,  2.,  3.,  2.,  2.,  2.,  3.,  3.,  3.,  
                   3.,  2.,  2.,  1.0])
```

```python
plt.plot(t, waveform)
plt.xlabel("Time (ns)",fontsize=20)
plt.ylabel("Intensity (bins)",fontsize=20)

plt.show()
```

```python
def func(x,A,B, mean,sig):
    return B+A*exp(-((x-mean)/sig)**2)
def LOL(x,A,A1,A2,B,m,m1,m2,sig,sig1,sig2):
    return func(x,A,B,m,sig)+A1*exp(-((x-m1)/sig1)**2)+A2*exp(-((x-m2)/sig2)**2)
```

```python
guess=array([27.,13.,8.,3.,17.,24.,28.,5.,3.,3.])
plt.plot(t, waveform,label="data")
plt.plot(t, LOL(t, guess[0],guess[1],guess[2],guess[3],guess[4],guess[5],guess[6],guess[7],guess[8],guess[8]),"r-",label="fit",linewidth=2)
plt.legend(loc="best")
plt.title("Initial Guess", fontsize=20)
plt.xlabel("Time (ns)",fontsize=20)
plt.ylabel("Intensity (bins)",fontsize=20)

plt.show()
```

```python

popt, pcov = curve_fit(LOL, t,waveform, guess)
print popt,pcov
```

```python
plt.plot(t, waveform,label="data")
plt.plot(t, LOL(t, popt[0],popt[1],popt[2],popt[3],popt[4],popt[5],popt[6],popt[7],popt[8],popt[8]),"r-",label="fit",linewidth=2)
plt.legend(loc="best")
plt.title("Curve Fitting", fontsize=20)
plt.xlabel("Time (ns)",fontsize=20)
plt.ylabel("Intensity (bins)",fontsize=20)

plt.show()
```
