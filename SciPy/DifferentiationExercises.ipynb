(a) Write two functions called twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) that take as arguments arrays of values x and y(x) and return the derivatives of y with respect to x. Make sure that the arrays returned by the function have the same shape as the input arrays.

```python
def twoPtForwardDiff(x,y):
    #Subtracts the difference from current value from the one in front of it 
    #for both x and y and return change y/ change x. (Forward difference).
    #Use backwards difference for last point
    dydx=np.zeros(y.shape,float)
    dydx[0:-1]=(y[1:]-y[:-1])/(x[1:]-x[:-1])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-y[-2])
    return dydx
def twoPtCenteredDiff(x,y):
    #Takes the difference between point behind and point in front for y and x. 
    #returns change y/change x. Last point uses forward backwards difference. 
    #first uses forward diff.
    dydx=np.zeros(y.shape,float)
    dydx[1:-1]=(y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0]=(y[1]-y[0])/(x[1]-x[0])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```


(b) Use your functions to take the derivative of the function y=8x4+3x over the range 0 <x< 10, with steps of 0.1. How do you know they work correctly? (Hint: compare to the analytical result).

```python
x=np.arange(0.0,10.0,0.1)
y=8*(x**4)+3*x
dydxEx=32*(x**3)+3
dydx1=twoPtForwardDiff(x,y)
dydx2=twoPtCenteredDiff(x,y)
plt.plot(x,dydx1,label='Forward difference')
plt.plot(x,dydx2,label='Center difference')
plt.plot(x,dydxEx,label='Exact')
plt.legend(loc='upper left')
plt.show()
```

(c) What is the accuracy of the results if you use a spacing of 0.01? (Plot the relative error)

```python
x=np.arange(0.0,10.0,0.01)
y=8*(x**4)+3*x
dydx1=twoPtForwardDiff(x,y)
dydx2=twoPtCenteredDiff(x,y)

dydxEx=32*(x**3)+3
per1=100*abs(dydx1[1:]-dydxEx[1:])/dydxEx[1:]
per2=100*abs(dydx2[1:]-dydxEx[1:])/dydxEx[1:]

plt.semilogy(x[1:],per1,label="forward diff")
plt.semilogy(x[1:],per2,label="Center diff")
plt.legend(loc="upper right")
```
