(a) Write a program that uses the Rejection Sampling Method to generate a set of 10,000 samples from this distribution and plot the density distribution, the CDF, and a histogram of the sampled distribution on the same graph.

```python
d=lambda x: 60.*(x**3)*(1.-x)**2
D=lambda x:10*x**6-24*x**5+15*x**4
x=np.linspace(0.,1.001,10000)
xmin,xmax=0.0,1.0
ymin,ymax=0.0,2.0736
#LOx = x[logical_and((x > xmin),(x < xmax))] Needed?
xsample=np.random.random_sample(10000)
ysample=ymax*np.random.random_sample(10000)
mask=(ysample<d(xsample))

plt.plot(x,d(x),lw=3, label="True distribution")
plt.plot(x,D(x),color='k',lw=3,label="CDF")
plt.xlim(-.3,1.3)
plt.plot([xmin, xmin], [ymin, ymax], color='b', linestyle='--')
plt.plot([xmax, xmax], [ymin, ymax], color='b', linestyle='--') 
plt.plot([xmin, xmax], [ymax, ymax], color='b', linestyle='--')
#plt.plot(xsample[:500],ysample[:500],'k.')
#plt.plot(xsample[mask][:500],ysample[mask][:500],'r.')
plt.hist(xsample[mask],bins=50,normed=True, label="Random Selected")
plt.legend(loc="best")
plt.show()
```
