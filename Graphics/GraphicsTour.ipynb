Common Formatting Tricks

```python
x=np.linspace(0,2*np.pi)
y1=np.sin(x)
y2=np.cos(x)

fig, ax1= plt.subplots(1,1)

ax1.set_title("My Plot", fontsize=20)
ax1.set_xlabel("x", fontsize=20)
ax1.set_ylabel("y", fontsize=20)

ax1.set_xlim([0,2])
ax1.set_ylim([-1,2])

ax1.plot(x,y1,'k--o', label="sin")
ax1.plot(x, y2, 'r-^', label="cos")

ax1.tick_params(direction='out', top=False, right=False)
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

ax1.annotate('(a)',(0.01, 0.96), size=12, xycoords='figure fraction')

ax1.legend()

fig.savefig('myfig-formatted.png')

```

Simple 3D plotting with Matlab

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01);
```

```python
from mpl_toolkits.mplot3d import Axes3D
dx=0.06
dy=0.06
x = np.arange(-2, 2, dx)
y = np.arange(-2, 2, dy)
x, y = np.meshgrid(x, y)

k = 1; q = 1; d = 1
V=(k*q/(x**2+(y-d)**2)**(1/2.)) - (k*q/(x**2+(y+d)**2)**(1/2.))

fig = figure()

ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth=0,cmap=cm.coolwarm)

cb = fig.colorbar(p, shrink=0.5)

xlabel("x")
ylabel("y")

savefig("Dipole.png")
```

Contour Plots

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
plt.contour(x,y,V,levels,linewidths=4,cmap=cm.coolwarm);
```
